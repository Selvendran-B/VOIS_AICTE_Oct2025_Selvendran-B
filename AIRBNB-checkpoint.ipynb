# DATA COLLECTION #

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display and plot styles
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid', palette='muted')
```

Load And Inspect Data

```python
file_path = r"C:\Users\ELCOT\Downloads\1730285881-Airbnb_Open_Data (1).xlsx"
df = pd.read_excel(file_path)
df.head()
```

```python
df.info()
```

Cell Basic Data Information

```python
# Check shape and info
print("Dataset Shape:", df.shape)
print("\nData Info:")
df.info()

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Quick summary of numeric columns
df.describe(include='all')
```

# DATA CLEANING #

```python
# Drop duplicated records
df.drop_duplicates(inplace=True)
print("After removing duplicates:", df.shape)
```

```python
# Drop 'house_rules' and 'license' columns
df.drop(columns=['house_rules', 'license'], inplace=True, errors='ignore')
print("After dropping columns:", df.shape)
```

```python
# Remove dollar signs and commas
for col in ['price', 'service fee']:
    if col in df.columns:
        df[col] = df[col].replace(r'[\$,]', '', regex=True).astype(float)

# Rename columns to include dollar sign for clarity
df.rename(columns={'price': 'price($)', 'service fee': 'service_fee($)'}, inplace=True)
```

```python
# Drop all rows with missing values
df.dropna(inplace=True)
print("After dropping missing values:", df.shape)
```

```python
df.rename(columns={'availability 365': 'availability_365'}, inplace=True)
```

```python
df['availability_365']
```

```python
# Remove outliers in 'availability_365' using IQR
Q1 = df['availability_365'].quantile(0.25)
Q3 = df['availability_365'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['availability_365'] >= lower_bound) & (df['availability_365'] <= upper_bound)]
print("After removing outliers:", df.shape)
```

# EXPLORATORY DATA ANALYSIS (EDA) #

```python
# Remove leading/trailing spaces and convert to lowercase
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('($)', '', regex=True)

# Check new column names
print(df.columns)
```

1---What are the different property types in the dataset?


```python
plt.figure(figsize=(10,5))
sns.countplot(
    data=df,
    y='room_type',
    order=df['room_type'].value_counts().index,
    palette='coolwarm'
)
plt.title('Distribution of Room Types')
plt.xlabel('Count')
plt.show()
```

Neighborhood group with the highest number of listings

```python
plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    x='neighbourhood_group',  # column name after cleaning
    order=df['neighbourhood_group'].value_counts().index,
    palette='viridis'
)
plt.title('Listings by Neighborhood Group')
plt.ylabel('Count')
plt.xlabel('Neighborhood Group')
plt.show()

# Print counts as table
df['neighbourhood_group'].value_counts()
```

Neighborhood group with highest average prices

```python
avg_prices = df.groupby('neighbourhood_group')['price($)'].mean().sort_values(ascending=False)
print("Average Price by Neighborhood Group:\n", avg_prices)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=avg_prices.index, y=avg_prices.values, palette='magma')
plt.title('Average Price by Neighborhood Group')
plt.ylabel('Average Price ($)')
plt.xlabel('Neighborhood Group')
plt.show()
```

Relationship between construction year and price

```python
plt.figure(figsize=(10,5))
sns.scatterplot(
    data=df,
    x='construction_year',  # cleaned column name
    y='price($)',
    alpha=0.5
)
plt.title('Price vs Construction Year')
plt.xlabel('Construction Year')
plt.ylabel('Price ($)')
plt.show()
```

Top 10 hosts by calculated host listing count

```python
top_hosts = df.groupby('host_name')['calculated_host_listings_count'].max().sort_values(ascending=False).head(10)
print("Top 10 Hosts by Listing Count:\n", top_hosts)

# Plot
plt.figure(figsize=(12,5))
sns.barplot(
    x=top_hosts.index,
    y=top_hosts.values,
    palette='cool'
)
plt.title('Top 10 Hosts by Listing Count')
plt.xticks(rotation=45)
plt.ylabel('Number of Listings')
plt.xlabel('Host Name')
plt.show()
```

Verified hosts and review rate

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x='host_identity_verified',
    y='review_rate_number',
    palette='Set2'
)
plt.title('Review Rate by Host Identity Verified')
plt.ylabel('Review Rate Number')
plt.xlabel('Host Identity Verified')
plt.show()
```

Correlation between price and service fee

```python
plt.figure(figsize=(10,5))
sns.scatterplot(
    data=df,
    x='price($)',
    y='service_fee($)',
    alpha=0.5
)
plt.title('Price vs Service Fee')
plt.xlabel('Price ($)')
plt.ylabel('Service Fee ($)')
plt.show()

# Calculate correlation
corr_price_fee = df[['price($)', 'service_fee($)']].corr().iloc[0,1]
print(f"Correlation between Price and Service Fee: {corr_price_fee:.2f}")
```

Average review rate by neighborhood group and room type

```python
avg_review = df.groupby(['neighbourhood_group', 'room_type'])['review_rate_number'].mean().unstack()
print("Average Review Rate:\n", avg_review)

plt.figure(figsize=(10,6))
sns.heatmap(
    avg_review,
    annot=True,
    fmt=".2f",
    cmap='YlGnBu'
)
plt.title('Average Review Rate by Neighborhood Group & Room Type')
plt.xlabel('Room Type')
plt.ylabel('Neighborhood Group')
plt.show()
```

Host listing count vs availability

```python
plt.figure(figsize=(10,5))
sns.scatterplot(
    data=df,
    x='calculated_host_listings_count',
    y='availability_365',
    alpha=0.5
)
plt.title('Host Listings Count vs Availability')
plt.xlabel('Calculated Host Listings Count')
plt.ylabel('Availability (days)')
plt.show()
```
